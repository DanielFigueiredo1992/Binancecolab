```python

import pandas as pd
import json
from google.colab import output
from datetime import datetime

def receive_data(data_json):
    global df
    df = pd.read_json(data_json)
    df.columns=[
        'open_time', 'open', 'high', 'low', 'close', 'volume',
        'close_time', 'quote_asset_volume', 'number_of_trades',
        'taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume', 'ignore'
      ]
    df['open_time'] = pd.to_datetime(df['open_time'], unit='ms')
    df['close_time'] = pd.to_datetime(df['close_time'], unit='ms')


output.register_callback('retornoDados', receive_data)

initial_datetime = datetime(2024, 5, 1, 0, 0, 0)
final_datetime = datetime.utcnow()
initial_epoch = int(initial_datetime.timestamp() * 1000)
final_epoch = int(final_datetime.timestamp() * 1000)
num_intervals = (final_epoch - initial_epoch) // (500 * 60000)

API_KEY = "m5gNycphXSKHEUT3yCOs875VqXzEGSudt1RGpML9O3PkxVCVNUODIUMiBe8PaTTi"
API_SECRET = "PRgpBJLfzWgnuyZ9OjfA2ODJk8v0YLyhyANXgtzYeud4b6dBGjMzM7EQDK4b5g0o"


from IPython.display import Javascript, display

# Exemplo de função para enviar dados do Python para o JavaScript
def sendparamtobinanceapi(data):
    # Converte o dicionário Python para uma string JSON
    data_json = json.dumps(data)

    # Usa a string JSON na string JavaScript
    display(Javascript("""
    (async function() {
        const data = """ + data_json + """;
        const initialEpoch = data.Rinitial_epoch;
        const finalEpoch = data.Rfinal_epoch;
        const numIntervals = data.Rnum_intervals;
        API_KEY = "m5gNycphXSKHEUT3yCOs875VqXzEGSudt1RGpML9O3PkxVCVNUODIUMiBe8PaTTi"
        API_SECRET = "PRgpBJLfzWgnuyZ9OjfA2ODJk8v0YLyhyANXgtzYeud4b6dBGjMzM7EQDK4b5g0o"
        // #'&apiKey=' + API_KEY + '&apiSecret=' + API_SECRET

        // Exibir os valores em uma janela de alerta
        async function getData(initialEpoch, finalEpoch, numIntervals) {
            let allData = [];
            let startTime = initialEpoch;
            let endTime = initialEpoch + 500 * 60000;
            for (let i = 0; i <= numIntervals; i++) {
                let url = 'https://api.binance.com/api/v3/klines?symbol=BTCUSDT&interval=1m&startTime=' + startTime + '&endTime=' + endTime;
                //let url = 'https://api.binance.com/api/v3/klines?symbol=BTCUSDT&interval=1m&startTime=' + startTime + '&endTime=' + endTime + '&apiKey=' + API_KEY + '&apiSecret=' + API_SECRET;
                //window.alert(url)
                let response = await fetch(url);
                let data = await response.json();

                allData.push(...data);

                startTime = endTime;
                endTime += 500 * 60000;
            }

            google.colab.kernel.invokeFunction('retornoDados', [JSON.stringify(allData)]);
        }

        await getData(initialEpoch, finalEpoch, numIntervals);
    })();
    """))



# Dados para enviar
data = {
    'Rinitial_epoch': initial_epoch,
    'Rfinal_epoch': final_epoch,
    'Rnum_intervals': num_intervals
}

# Envie os dados para o JavaScript
sendparamtobinanceapi(data)
```
